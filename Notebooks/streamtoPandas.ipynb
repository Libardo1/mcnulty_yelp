```python
import json
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read json files from tweeter
tweets_data_path = 'fetched_tweets.txt'

tweets_data = []
with open(tweets_data_path, "r") as tweets_file:
    for line in tweets_file:
        try:
            tweet = json.loads(line)
            tweets_data.append(tweet)
        except:
            continue
```

```python
# How many tweets we have?
len(tweets_data)
```

```python
# Create list for columns
track=['Cheesecake', 'olivegarden', 'Chilis', 'redlobster', 'Outback', 'TGIFridays', 'PFChangs', 'LongHornSteaks', 'Hooters', 'calpizzakitchen', 'bjsrestaurants', 'MacaroniGrill', 'LogansRoadhouse']
text = []
created_at = []
lang = []
location = []
user_id = []
business = []

for restaurant in track:
    for i in range(0,len(tweets_data)):
        try:
            if (tweets_data[i]['entities']['user_mentions'][0]['screen_name']) == restaurant:
                business.append(restaurant)
                print (i, tweets_data[i]['id'])
                user_id.append(tweets_data[i]['id'])
                
                print (i, tweets_data[i]['text'])
                text.append(tweets_data[i]['text'])
            
                print (i, tweets_data[i]['created_at'])
                created_at.append(tweets_data[i]['created_at'])
            
                print (i, tweets_data[i]['lang'])
                lang.append(tweets_data[i]['lang'])
            
                if (tweets_data[i]['user']['location'] == None):
                    print (i, np.nan)
                    location.append(np.nan)
                else:
                    print (i, tweets_data[i]['user']['location'])
                    location.append(tweets_data[i]['user']['location'])
                print ('')
        except:
            continue
```

```python
# Create a dictionary
mentions = {}
mentions['text'] = text
mentions['created_at'] = created_at
mentions['user_id'] = user_id
mentions['lang'] = lang
mentions['location'] = location
mentions['business'] = business
```

```python
# Convert dictiona
df = DataFrame(mentions)
```

```python
# Replace https...with empty string
def http(text):
    if re.search('https?:\/\/.*[\r\n]*', text):
        return re.sub('https?:\/\/.*[\r\n]*', '', text)
    else:
        return text
```

```python
df4["text_no_https"] = df4.text.apply(http)
```

```python
# Replace @.... with empty string
def at(text):
    if re.search('@([\w.-]+)', text):
        return re.sub('@([\w.-]+)', '', text)
    else:
        return text
```

```python
df4["text_no_ats"] = df4['text_no_https'].apply(at)
```
